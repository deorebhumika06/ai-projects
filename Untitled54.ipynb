```python
import pandas as pd

# Example data
data = [2, 4, 6, 8, 10, 50]

df = pd.DataFrame(data, columns=["values"])

Q1 = df["values"].quantile(0.25)
Q3 = df["values"].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

# Check outliers
outliers = df[(df["values"] < lower) | (df["values"] > upper)]
print(outliers)
```

```python
import pandas as pd
df = pd.read_csv('/content/indian_ecommerce_pricing_revenue_growth_36_months (1).csv')
numeric_df = df.select_dtypes(include=['number'])
Q1 = numeric_df.quantile(0.25)
Q3 = numeric_df.quantile(0.75)
IQR = Q3 - Q1

# Calculate lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Create a boolean mask for rows that are not outliers in any numeric column
outlier_mask = ((numeric_df < lower_bound) | (numeric_df > upper_bound)).any(axis=1)

# Filter the original DataFrame to keep only non-outlier rows
df_no_outliers = df[~outlier_mask]
```

```python
# Display the detected outliers
outliers_df = df[outlier_mask]
print(f"Number of outliers detected: {len(outliers_df)}")
outliers_df.head()
```
